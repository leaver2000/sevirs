```python
# sevir dataset constants
MINUTES_PER_HOUR = 60
HOURS_PER_SAMPLE = 4
IMAGE_TIME_INTERVAL = 5
#
TIME_SAMPLES = ((MINUTES_PER_HOUR * HOURS_PER_SAMPLE) // IMAGE_TIME_INTERVAL) + 1
#
ONE_HOUR_OF_SAMPLES = MINUTES_PER_HOUR // IMAGE_TIME_INTERVAL

IMG_IDS = ("a", "b")


def time_series_generator(
    n_inputs: int = ONE_HOUR_OF_SAMPLES * 2,  # two hours of samples
    n_targets: int = ONE_HOUR_OF_SAMPLES,  # one hour of samples
    step=1,  # step size between each sample
):
    # n_inputs = self.n_inputs
    # n_targets = self.n_outputs
    # step = self.step
    for id_ in IMG_IDS:
        # arr = self.select(self.img_ids[idx])
        for x_start in range(0, TIME_SAMPLES, step):
            x_stop = x_start + n_inputs
            y_start = x_stop + 1
            y_stop = y_start + n_targets
            if y_stop > TIME_SAMPLES:
                continue
            yield (
                f"""[{id_=}]
(
    tensor[C, X, Y, {x_start}:{x_stop}],
    tensor[C, X, Y, {y_start}:{y_stop}]
)\n"""
            )


data = list(time_series_generator())
print("\n".join(data))
# 60//5
```
